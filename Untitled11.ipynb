```python
def rearrange_max_min(arr):
    n = len(arr)
    max_idx = n - 1
    min_idx = 0
    max_elem = arr[-1] + 1

    for i in range(n):
        if i % 2 == 0:
            arr[i] += (arr[max_idx] % max_elem) * max_elem
            max_idx -= 1
        else:
            arr[i] += (arr[min_idx] % max_elem) * max_elem
            min_idx += 1

    for i in range(n):
        arr[i] //= max_elem

    return arr


# Example 1
arr1 = [1, 2, 3, 4, 5, 6, 7]
print("Original array:", arr1)
rearranged1 = rearrange_max_min(arr1)
print("Rearranged array:", rearranged1)

# Example 2
arr2 = [1, 2, 3, 4, 5, 6]
print("\nOriginal array:", arr2)
rearranged2 = rearrange_max_min(arr2)
print("Rearranged array:", rearranged2)
```

```python
def findCircleNum(isConnected):
    n = len(isConnected)
    visited = [False] * n

    def dfs(city):
        for neighbor in range(n):
            if isConnected[city][neighbor] == 1 and not visited[neighbor]:
                visited[neighbor] = True
                dfs(neighbor)

    provinces = 0
    for city in range(n):
        if not visited[city]:
            visited[city] = True
            dfs(city)
            provinces += 1

    return provinces


# Example
isConnected = [
    [1, 1, 0],
    [1, 1, 0],
    [0, 0, 1]
]

print("Output:", findCircleNum(isConnected))
```

```python
def stringSubstrings(words):
    result = []

    for i in range(len(words)):
        for j in range(len(words)):
            if i != j and words[i] in words[j]:
                result.append(words[i])
                break

    return result


# Example 1
words1 = ["mass", "as", "hero", "superhero"]
print("Output:", stringSubstrings(words1))

# Example 2
words2 = ["blue", "green", "bu", "red"]
print("Output:", stringSubstrings(words2))

# Example 3
words3 = ["lemon", "melon", "le", "on"]
print("Output:", stringSubstrings(words3))

# Example 4
words4 = ["apple", "pie", "applepie", "pi"]
print("Output:", stringSubstrings(words4))
```
