```python
def find_subarray_with_sum(arr, S):
    n = len(arr)
    start = 0
    curr_sum = 0
    for end in range(n):
        curr_sum += arr[end]
        while curr_sum > S and start <= end:
            curr_sum -= arr[start]
            start += 1
        if curr_sum == S:
            return start + 1, end + 1
    return -1

A = [4, 2, 0, 1, 6]
S = 3
print(find_subarray_with_sum(A, S))
A = [5, 8, 12, 7, 6, 2]
S = 25
print(find_subarray_with_sum(A, S))
A = [10, 20, 30, 40, 50]
S = 100
print(find_subarray_with_sum(A, S))
A = [1, 1, 1, 1, 1, 1]
S = 4
print(find_subarray_with_sum(A, S))
A = [7, 3, 2, 5, 8, 1]
S = 10
print(find_subarray_with_sum(A, S))
```

```python
def findTriplets(arr, n):
    arr.sort()
    for i in range(n-2):
        l = i + 1
        r = n - 1
        while l < r:
            s = arr[i] + arr[l] + arr[r]
            if s == 0:
                return 1
            elif s < 0:
                l += 1
            else:
                r -= 1
    return 0

print(findTriplets([0, -1, 2, -3, 1], 5))
print(findTriplets([1, 2, 3], 3))
```

```python
def dfs(node, graph, visited, stack):
    visited[node] = True
    stack[node] = True
    for nxt in graph[node]:
        if not visited[nxt]:
            if dfs(nxt, graph, visited, stack):
                return True
        elif stack[nxt]:
            return True
    stack[node] = False
    return False

def hasCycle(graph, V):
    visited = [False] * V
    stack = [False] * V
    for i in range(V):
        if not visited[i]:
            if dfs(i, graph, visited, stack):
                return True
    return False

V = 4
graph = [[] for _ in range(V)]
graph[0].append(1)
graph[1].append(2)
graph[2].append(3)
graph[3].append(0)
if hasCycle(graph, V):
    print("Cycle is Found")
else:
    print("No Cycle Found")
```

```python
def findStartingStation(gas, cost):
    total = 0
    start = 0
    tank = 0
    for i in range(len(gas)):
        diff = gas[i] - cost[i]
        tank += diff
        total += diff
        if tank < 0:
            start = i + 1
            tank = 0
    return start if total >= 0 else -1

print(findStartingStation([1,2,3,4,5], [3,4,5,1,2]))
print(findStartingStation([2,3,4], [3,4,3]))
print(findStartingStation([3,3,4], [3,4,3]))
print(findStartingStation([0,4,3,2], [2,2,2,2]))
```

```python
def checkBalanced(exp):
    stack = []
    for ch in exp:
        if ch in "({[":
            stack.append(ch)
        elif ch in ")}]":
            if not stack:
                return "Not Balanced"
            top = stack.pop()
            if (ch == ")" and top != "(") or (ch == "}" and top != "{") or (ch == "]" and top != "["):
                return "Not Balanced"
    if stack:
        return "Not Balanced"
    else:
        return "Balanced"

print(checkBalanced("[()]{}{[()()]()}"))
print(checkBalanced("[(])"))
```

```python
def LCS(text1, text2):
    n = len(text1)
    m = len(text2)
    dp = [[0]*(m+1) for _ in range(n+1)]
    for i in range(1, n+1):
        for j in range(1, m+1):
            if text1[i-1] == text2[j-1]:
                dp[i][j] = 1 + dp[i-1][j-1]
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    return dp[n][m]

print(LCS("abcde", "ace"))
print(LCS("abc", "abc"))
print(LCS("abc", "def"))
```

```python
def solve(arr, m):
    def check(maxp):
        s = 1
        p = 0
        for x in arr:
            if x > maxp:
                return False
            if p + x > maxp:
                s += 1
                p = x
            else:
                p += x
        return s <= m
    if m > len(arr):
        return -1
    l = max(arr)
    r = sum(arr)
    ans = r
    while l <= r:
        mid = (l + r) // 2
        if check(mid):
            ans = mid
            r = mid - 1
        else:
            l = mid + 1
    return ans

print(solve([12, 34, 67, 90], 2))
print(solve([15, 17, 20], 2))
```

```python
def coinChange(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for i in range(1, amount + 1):
        for coin in coins:
            if coin <= i:
                dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

print(coinChange([25, 10, 5], 30))
print(coinChange([9, 6, 5, 1], 19))
print(coinChange([5, 1], 0))
print(coinChange([4, 6, 2], 5))
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def hasCycle(head):
    if not head or not head.next:
        return False
    slow = head
    fast = head.next
    while fast and fast.next:
        if slow == fast:
            return True
        slow = slow.next
        fast = fast.next.next
    return False

node1 = ListNode(1)
node3 = ListNode(3)
node4 = ListNode(4)
node1.next = node3
node3.next = node4
node4.next = node3
print(hasCycle(node1))
node1 = ListNode(1)
node8 = ListNode(8)
node3 = ListNode(3)
node4 = ListNode(4)
node1.next = node8
node8.next = node3
node3.next = node4
node4.next = None
print(hasCycle(node1))
```

```python
def foldMatrix(matrix, n):
    for i in range(n // 2):
        for j in range(n):
            matrix[i][j] += matrix[n - 1 - i][j]
    if n % 2 == 0:
        rows = n // 2
    else:
        rows = n // 2 + 1
    for i in range(rows):
        for j in range(n // 2):
            matrix[i][j] += matrix[i][n - 1 - j]
    if n % 2 == 0:
        cols = n // 2
    else:
        cols = n // 2 + 1
    result = []
    for i in range(rows):
        row = []
        for j in range(cols):
            row.append(matrix[i][j])
        result.append(row)
    return result

n = 4
matrix = [[1, 0, 2, 4],
          [-1, 5, 3, 1],
          [8, 2, 6, 7],
          [10, 9, 3, 8]]
result = foldMatrix(matrix, n)
for row in result:
    print(*row)
print()
n = 3
matrix = [[1, 0, 2],
          [-1, 5, 3],
          [8, 2, 6]]
result = foldMatrix(matrix, n)
for row in result:
    print(*row)
```

```python
def maxGuests(arr, exit):
    events = []
    for i in range(len(arr)):
        events.append((arr[i], 1))
        events.append((exit[i], -1))
    events.sort()
    maxCount = 0
    currentCount = 0
    maxTime = 0
    for time, event in events:
        currentCount += event
        if currentCount > maxCount:
            maxCount = currentCount
            maxTime = time
    return maxTime

arr = [1, 2, 9, 5, 5]
exit = [4, 5, 12, 9, 12]
print(maxGuests(arr, exit))
```

```python
def findPeak(arr):
    n = len(arr)
    if n == 1:
        return arr[0]
    if arr[0] >= arr[1]:
        return arr[0]
    if arr[n-1] >= arr[n-2]:
        return arr[n-1]
    for i in range(1, n-1):
        if arr[i] >= arr[i-1] and arr[i] >= arr[i+1]:
            return arr[i]

arr1 = [5, 10, 20, 15]
print(findPeak(arr1))
arr2 = [10, 20, 15, 2, 23, 90, 67]
print(findPeak(arr2))
```

```python
def countWays(N):
    MOD = 1000000007

    if N == 0:
        return 1
    elif N == 1:
        return 1
    elif N == 2:
        return 2

    dp = [0] * (N + 1)
    dp[0], dp[1], dp[2] = 1, 1, 2

    for i in range(3, N + 1):
        dp[i] = (dp[i-1] + dp[i-2] + dp[i-3]) % MOD

    return dp[N] % MOD

N = int(input("Enter the number of steps (N): "))
print("Number of ways to reach step", N, "=", countWays(N))
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def countNodes(head):
    count = 0
    while head:
        count += 1
        head = head.next
    return count

globalHead = None

def sortedListToBSTHelper(n):
    global globalHead
    if n <= 0:
        return None

    left = sortedListToBSTHelper(n // 2)

    root = TreeNode(globalHead.val)
    root.left = left

    globalHead = globalHead.next

    root.right = sortedListToBSTHelper(n - n // 2 - 1)
    return root

def sortedListToBST(head):
    global globalHead
    globalHead = head
    n = countNodes(head)
    return sortedListToBSTHelper(n)

def printTree(root, level=0):
    if root is not None:
        printTree(root.right, level + 1)
        print(' ' * 4 * level + str(root.val))
        printTree(root.left, level + 1)

head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)

root = sortedListToBST(head)

print("Input: Linked List 1->2->3->4")
print("Output: A Balanced BST (as per example)\n")
printTree(root)
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.prev = None
        self.next = None
        self.child = None

def flatten(head):
    if not head:
        return None

    stack = []
    curr = head

    while curr:
        if curr.child:
            if curr.next:
                stack.append(curr.next)
            curr.next = curr.child
            curr.child.prev = curr
            curr.child = None

        # If reached end and stack not empty, pop and connect
        if not curr.next and stack:
            nxt = stack.pop()
            curr.next = nxt
            nxt.prev = curr

        curr = curr.next

    return head

def printList(head):
    while head:
        print(head.val, end=" ")
        head = head.next
    print()

head = Node(1)
head.next = Node(2); head.next.prev = head
head.next.next = Node(3); head.next.next.prev = head.next
head.next.next.next = Node(4); head.next.next.next.prev = head.next.next

head.next.child = Node(5)
head.next.child.next = Node(6); head.next.child.next.prev = head.next.child
head.next.child.next.child = Node(7)
head.next.child.next.child.next = Node(8); head.next.child.next.child.next.prev = head.next.child.next.child

flat_head = flatten(head)

# Output
print("Flattened list:")
printList(flat_head)
```

```python
def maxProfit(prices, fee):
    cash = 0
    hold = -prices[0]

    for price in prices[1:]:
        cash = max(cash, hold + price - fee)
        hold = max(hold, cash - price)

    return cash

# Example 1
prices = [1,3,2,8,4,9]
fee = 2
print("Output:", maxProfit(prices, fee))

# Example 2
prices = [1,3,7,5,10,3]
fee = 3
print("Output:", maxProfit(prices, fee))
```

```python
def cutRod(price, N):
    dp = [0] * (N + 1)

    for i in range(1, N + 1):
        for j in range(1, i + 1):
            dp[i] = max(dp[i], price[j - 1] + dp[i - j])

    return dp[N]

# Example 1
N = 8
price = [1, 5, 8, 9, 10, 17, 17, 20]
print("Output:", cutRod(price, N))

# Example 2
N = 1
price = [3]
print("Output:", cutRod(price, N))
```

```python
def minCost(str1, str2):
    n = len(str1)
    cost = 0
    i = 0

    while i < n:
        if str1[i] != str2[i]:
            if i + 1 < n and str1[i + 1] != str2[i + 1] and str1[i] != str1[i + 1]:
                cost += 1
                i += 2
            else:
                cost += 1
                i += 1
        else:
            i += 1

    return cost

# Example 1
str1 = "abb"
str2 = "baa"
print("Output:", minCost(str1, str2))

# Example 2
str1 = "abab"
str2 = "aabb"
print("Output:", minCost(str1, str2))
```

```python
# Definition for a binary tree node
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Function to convert BST to Greater Tree
def convertBST(root):
    total = 0

    def reverseInorder(node):
        nonlocal total
        if not node:
            return
        # Visit right subtree first
        reverseInorder(node.right)
        # Update current node
        total += node.val
        node.val = total
        # Visit left subtree
        reverseInorder(node.left)

    reverseInorder(root)
    return root

# Helper function to print tree in level order
from collections import deque
def levelOrder(root):
    if not root:
        return []
    result = []
    q = deque([root])
    while q:
        node = q.popleft()
        if node:
            result.append(node.val)
            q.append(node.left)
            q.append(node.right)
        else:
            result.append(None)
    # Remove trailing None values for cleaner output
    while result and result[-1] is None:
        result.pop()
    return result


# ---------------------------
# Example tree construction
# Input: root = [4,1,6,0,2,5,7,None,None,None,3,None,None,None,8]
# ---------------------------

root = TreeNode(4)
root.left = TreeNode(1)
root.right = TreeNode(6)
root.left.left = TreeNode(0)
root.left.right = TreeNode(2)
root.left.right.right = TreeNode(3)
root.right.left = TreeNode(5)
root.right.right = TreeNode(7)
root.right.right.right = TreeNode(8)

# Convert BST to Greater Tree
newRoot = convertBST(root)

# Output in level order (like example)
print("Output:", levelOrder(newRoot))
```

```python
# Definition for a binary tree node (Department)
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Function to compute maximum allocation value
def maxAllocation(root):
    def dfs(node):
        if not node:
            return (0, 0)  # (include, exclude)

        left = dfs(node.left)
        right = dfs(node.right)

        # If we include this node → we cannot include its children
        include = node.val + left[1] + right[1]

        # If we exclude this node → we can choose to include or exclude its children
        exclude = max(left) + max(right)

        return (include, exclude)

    return max(dfs(root))

# ----------------------------
# Example Tree Construction
# CEO (10)
#     /      \
#  Finance(1)  Sales(2)
#   /    \         \
# HR(4)  IT(5)   Marketing(3)
# ----------------------------

root = TreeNode(10)
root.left = TreeNode(1)
root.right = TreeNode(2)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.right = TreeNode(3)

# Calculate maximum allocation
print("Output:", maxAllocation(root))
```

```python
from collections import defaultdict, deque

# Function to build the tree and find the root (CEO)
def build_tree(n, relationships):
    tree = defaultdict(list)
    indegree = [0] * (n + 1)

    for manager, emp in relationships:
        tree[manager].append(emp)
        indegree[emp] += 1

    # Root (CEO) has no manager
    root = None
    for i in range(1, n + 1):
        if indegree[i] == 0:
            root = i
            break

    return tree, root

# Function to find LCA (Closest Common Manager)
def findLCA(root, emp1, emp2, tree):
    if not root:
        return None
    if root == emp1 or root == emp2:
        return root

    found = []
    for child in tree[root]:
        res = findLCA(child, emp1, emp2, tree)
        if res:
            found.append(res)

    # If both employees are found in different subtrees
    if len(found) == 2:
        return root
    # Otherwise, return whichever subtree has them
    elif len(found) == 1:
        return found[0]
    else:
        return None

# ----------------------------
# Example Test Case
# ----------------------------

n = 8
relationships = [
    (1, 2),
    (1, 3),
    (1, 4),
    (2, 5),
    (2, 6),
    (4, 7),
    (4, 8)
]
emp1, emp2 = 5, 8

# Build the tree
tree, root = build_tree(n, relationships)

# Find closest common manager
manager = findLCA(root, emp1, emp2, tree)

print(f"Closest common manager for {emp1} and {emp2}: {manager}")
```

```python
def rearrange(arr):
    n = len(arr)
    max_idx = n - 1
    min_idx = 0
    max_elem = arr[-1] + 1  # one greater than the max element

    for i in range(n):
        if i % 2 == 0:
            arr[i] += (arr[max_idx] % max_elem) * max_elem
            max_idx -= 1
        else:
            arr[i] += (arr[min_idx] % max_elem) * max_elem
            min_idx += 1

    # Final decoding to get rearranged array
    for i in range(n):
        arr[i] //= max_elem

    return arr

# Example 1
arr = [1, 2, 3, 4, 5, 6, 7]
print("Output:", rearrange(arr))  # Expected: [7, 1, 6, 2, 5, 3, 4]

# Example 2
arr = [1, 2, 3, 4, 5, 6]
print("Output:", rearrange(arr))  # Expected: [6, 1, 5, 2, 4, 3]
```

```python
def findCircleNum(isConnected):
    n = len(isConnected)
    visited = [False] * n

    def dfs(i):
        for j in range(n):
            if isConnected[i][j] == 1 and not visited[j]:
                visited[j] = True
                dfs(j)

    provinces = 0
    for i in range(n):
        if not visited[i]:
            dfs(i)
            provinces += 1

    return provinces

# Example 1
isConnected = [
    [1, 1, 0],
    [1, 1, 0],
    [0, 0, 1]
]
print("Output:", findCircleNum(isConnected))  # Expected Output: 2
```

```python
def stringSubstrings(words):
    result = []
    n = len(words)

    for i in range(n):
        for j in range(n):
            if i != j and words[i] in words[j]:
                result.append(words[i])
                break  # Found at least one match, no need to check further
    return result

# Example 1
words = ["mass", "as", "hero", "superhero"]
print("Output:", stringSubstrings(words))  # Expected: ["as", "hero"]

# Example 2
words = ["blue", "green", "bu", "red"]
print("Output:", stringSubstrings(words))  # Expected: []

# Example 3
words = ["lemon", "melon", "le", "on"]
print("Output:", stringSubstrings(words))  # Expected: ["le", "on"]

# Example 4
words = ["apple", "pie", "applepie", "pi"]
print("Output:", stringSubstrings(words))  # Expected: ["apple", "pie", "pi"]
```
